Text Summarization using Gensim
The summarize( ) function implements the text rank summarization.
You do not have to generate a tokenized list by splitting the sentences as that is already handled by the gensim.summarization.textcleaner module. 

```python
# Sample text content
sample_text = """
Artificial Intelligence (AI) is rapidly transforming the world. From healthcare to transportation, 
AI technologies are enabling smarter decisions and automation. In healthcare, AI algorithms assist 
in disease detection and personalized medicine. In transportation, self-driving vehicles are becoming 
more reliable thanks to AI. Moreover, businesses use AI for predictive analytics and customer engagement.

However, AI also raises important ethical concerns. Issues like bias in algorithms, data privacy, and 
job displacement must be addressed responsibly. Governments and institutions are forming regulations 
to ensure AI benefits society as a whole.

As AI continues to evolve, collaboration between technologists, policymakers, and ethicists is crucial. 
The future of AI depends not just on technical advancement but also on how responsibly it is deployed 
in real-world scenarios.
"""

# Write to sample_data.txt
with open('sample_data.txt', 'w', encoding='utf-8') as f:
    f.write(sample_text)

print("sample_data.txt has been written.")
```

```python
from gensim.summarization import summarize

with open('sample_data.txt', 'r', encoding='utf-8') as f:
    text = f.read()

print("Summary:")
print(summarize(text, word_count=25))
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer

with open('sample_data.txt', 'r', encoding='utf-8') as f:
    text = f.read()

parser = PlaintextParser.from_string(text, Tokenizer("english"))
summarizer = LsaSummarizer()
summary = summarizer(parser.document, 3)  # 3 sentences

for sentence in summary:
    print(sentence)
```

```python
from gensim.summarization.summarizer import summarize
from gensim.summarization import keywords
import wikipedia
import en_core_web_sm

# Get wiki content.
wikisearch = wikipedia.page("Amitabh Bachchan")
wikicontent = wikisearch.content
nlp = en_core_web_sm.load()
doc = nlp(wikicontent)

# Save the wiki content to a file
# (for reference).
f = open("wikicontent.txt", "w")
f.write(wikicontent)
f.close()

# Summary (0.5% of the original content).
summ_per = summarize(wikicontent, ratio = 0.05)
print("Percent summary")
print(summ_per)

# Summary (200 words)
summ_words = summarize(wikicontent, word_count = 200)
print("Word count summary")
print(summ_words)
```

```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
import wikipedia

# Get wiki content
wikisearch = wikipedia.page("Amitabh Bachchan")
wikicontent = wikisearch.content

# Save the content
with open("wikicontent.txt", "w", encoding="utf-8") as f:
    f.write(wikicontent)

# Summarize
parser = PlaintextParser.from_string(wikicontent, Tokenizer("english"))
summarizer = LsaSummarizer()

summary = summarizer(parser.document, 5)  # 5 sentences
for sentence in summary:
    print(sentence)
```

| 🧱 **Heavy Model/Library**         | Libraries like Hugging Face Transformers load large models that need **Torch or TensorFlow** and high memory.                |
| 🔧 **Missing Native Dependencies** | Some libraries (like `gensim.summarization`, `transformers`, `spacy`) depend on native C extensions or proper installations. |
| ❌ **Conflicting/Corrupt Env**      | Conda environment has conflicting or improperly installed packages.                                                          |
| 🔌 **GPU/Torch Errors**            | If `torch` is installed but GPU drivers (like CUDA) are missing/mismatched, kernel can crash.                                |
