```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

def tsne_plot(word_vectors, vocab):
    labels = []
    wordvecs = []

    for word in vocab:
        if word in word_vectors:
            wordvecs.append(word_vectors[word])
            labels.append(word)

    wordvecs = np.array(wordvecs)  # Convert to NumPy array for TSNE

    if len(wordvecs) == 0:
        print("No word vectors found.")
        return

    if len(wordvecs) <= 3:
        print(f"Not enough data points for t-SNE (got {len(wordvecs)}). Increase vocab size.")
        return

    tsne_model = TSNE(perplexity=5, n_components=2, init='pca', random_state=42)
    coordinates = tsne_model.fit_transform(wordvecs)

    x = [val[0] for val in coordinates]
    y = [val[1] for val in coordinates]

    plt.figure(figsize=(10, 10))
    for i in range(len(x)):
        plt.scatter(x[i], y[i])
        plt.annotate(labels[i], xy=(x[i], y[i]), ha='right')
    plt.show()


```

The t-SNE plot typically shows words as points on a 2D space to visualize their semantic similarity, but does not include arrows by default, because:

ðŸ” Why There Are No Arrows in t-SNE:
t-SNE is for clustering/embedding, not direction or flow.

The coordinates are non-linear projections, so vector directions lose their meaning.

Arrows (vectors) would imply directionality in semantic relationships â€” but t-SNE distorts distances to maintain local similarities, not global direction.

If You Want Arrows:
You can manually add arrows between specific word pairs to indicate relationships (e.g., "king" âž "queen").

Here's an example of how to draw arrows between selected word pairs:

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

def tsne_plot_with_arrows(model, word_pairs):
    words = list(set([w for pair in word_pairs for w in pair]))
    word_vecs = [model.wv[w] for w in words]

    tsne = TSNE(n_components=2, init='pca', random_state=42,perplexity=min(30, len(words) - 1))
    coords = tsne.fit_transform(np.array(word_vecs))
    word_coords = {word: coord for word, coord in zip(words, coords)}

    plt.figure(figsize=(10, 8))

    for word, (x, y) in word_coords.items():
        plt.scatter(x, y)
        plt.text(x + 0.5, y, word, fontsize=12)

    for w1, w2 in word_pairs:
        x1, y1 = word_coords[w1]
        x2, y2 = word_coords[w2]
        plt.arrow(x1, y1, x2 - x1, y2 - y1, color='red', head_width=0.3, length_includes_head=True)

    plt.title("t-SNE with Word Vectors and Arrows")
    plt.show()

# Example usage with some semantically related word pairs
#word_pairs=[('america', 'pay'), ('north_america', 'universities')]
tsne_plot_with_arrows(model,[('america', 'pay'), ('north_america', 'universities')])
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

def tsne_plot_with_arrows(model, word_pairs):
    # Filter valid pairs
    valid_pairs = []
    missing_words = set()

    for w1, w2 in word_pairs:
        if w1 in model.wv.key_to_index and w2 in model.wv.key_to_index:
            valid_pairs.append((w1, w2))
        else:
            if w1 not in model.wv.key_to_index:
                missing_words.add(w1)
            if w2 not in model.wv.key_to_index:
                missing_words.add(w2)

    if missing_words:
        print("Warning: These words were not in the vocabulary and will be skipped:", missing_words)
    if not valid_pairs:
        print("No valid word pairs found in vocabulary.")
        return

    words = list(set([w for pair in valid_pairs for w in pair]))
    word_vecs = [model.wv[w] for w in words]

    tsne = TSNE(n_components=2, init='pca', random_state=42,perplexity=min(30, len(words) - 1))#perplexity (30.0) must be less than n_samples (4)
    coords = tsne.fit_transform(np.array(word_vecs))
    word_coords = {word: coord for word, coord in zip(words, coords)}

    plt.figure(figsize=(10, 8))
    for word, (x, y) in word_coords.items():
        plt.scatter(x, y)
        plt.text(x + 0.5, y, word, fontsize=12)

    for w1, w2 in valid_pairs:
        x1, y1 = word_coords[w1]
        x2, y2 = word_coords[w2]
        plt.arrow(x1, y1, x2 - x1, y2 - y1, color='red', head_width=0.2, length_includes_head=True)

    plt.title("t-SNE with Word Vectors and Arrows")
    plt.show()
```

```python
# Visualize t-SNE for the top N most frequent words in your Word2Vec model â€” with optional arrows between semantically related words (if any).
#Step 1: Function to Extract Top-N Frequent Words

def get_top_n_words(model, n=50):
    return [word for word, _ in model.wv.index_to_key[:n]]
#This uses the .index_to_key property (available in Gensim 4+) to get the most frequent words.

# Step 2: Plot t-SNE for Top-N Words (with optional arrows)

from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

def tsne_top_n_words(model, top_n=50, word_pairs=None):
    words = get_top_n_words(model, n=top_n)
    word_vecs = [model.wv[w] for w in words]

    tsne = TSNE(n_components=2, init='pca', random_state=42, perplexity=min(30, len(words) - 1))
    coords = tsne.fit_transform(np.array(word_vecs))
    word_coords = {word: coord for word, coord in zip(words, coords)}

    plt.figure(figsize=(12, 10))

    for word, (x, y) in word_coords.items():
        plt.scatter(x, y)
        plt.text(x + 0.5, y, word, fontsize=10)

    # Optional: draw arrows between selected related word pairs
    if word_pairs:
        valid_pairs = [(w1, w2) for w1, w2 in word_pairs if w1 in word_coords and w2 in word_coords]
        for w1, w2 in valid_pairs:
            x1, y1 = word_coords[w1]
            x2, y2 = word_coords[w2]
            plt.arrow(x1, y1, x2 - x1, y2 - y1, color='red', head_width=0.2, length_includes_head=True)

    plt.title(f"t-SNE of Top {top_n} Word Vectors")
    plt.show()
```

```python

#If you're just trying to get the top N words, this line should simply be:
def get_top_n_words(model, n=50):
    return model.wv.index_to_key[:n]

def tsne_top_n_words(model, top_n=50, word_pairs=None):
    words = get_top_n_words(model, n=top_n)
    word_vecs = np.array([model.wv[w] for w in words])

    tsne = TSNE(n_components=2, init='pca', random_state=42, perplexity=min(30, len(words) - 1))
    reduced_vecs = tsne.fit_transform(word_vecs)

    plt.figure(figsize=(12, 8))
    for i, word in enumerate(words):
        plt.scatter(reduced_vecs[i, 0], reduced_vecs[i, 1])
        plt.annotate(word, (reduced_vecs[i, 0], reduced_vecs[i, 1]))
    plt.title(f"t-SNE visualization of top {top_n} words")
    plt.show()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def generate_similar_word_pairs(model, top_n=50, top_k=1):
    words = get_top_n_words(model, top_n)
    word_vecs = np.array([model.wv[word] for word in words])     

    sim_matrix = cosine_similarity(word_vecs)
    np.fill_diagonal(sim_matrix, -np.inf)  # avoid self-matching

    pairs = set()
    for i, word in enumerate(words):
        most_similar_indices = np.argsort(sim_matrix[i])[-top_k:]
        for j in most_similar_indices:
            pairs.add((word, words[j]))

    return list(pairs)
#top_k: how many related words you want per word (1 = most similar)

#Step 2: Use These Pairs in Your t-SNE Plot
# Generate the word pairs
word_pairs = generate_similar_word_pairs(model, top_n=50, top_k=1)

# Plot with arrows showing relationships
tsne_top_n_words(model, top_n=50, word_pairs=word_pairs)
```
